## Load at a particular day

```python
import matplotlib.pyplot as plt

# Convert the 'DateTime' column to datetime format
data2018['DateTime'] = pd.to_datetime(data2018['DateTime'])

# Choose the date you want to plot (e.g., '2018-01-01')
selected_date = '2018-09-13'

# Filter the DataFrame for the selected date
selected_day_data = data2018[data2018['DateTime'].dt.date == pd.to_datetime(selected_date).date()]

# Plot the 'Load' column for the selected day
plt.figure(figsize=(18, 10))  # Set the figure size
plt.plot(selected_day_data['DateTime'], selected_day_data['Load'], label='Load')
plt.xlabel('Time')
plt.ylabel('Load')
plt.title(f'Load for {selected_date}')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert the 'DateTime' column to datetime format
data2018['DateTime'] = pd.to_datetime(data2018['DateTime'])
data2019['DateTime'] = pd.to_datetime(data2019['DateTime'])
data2020['DateTime'] = pd.to_datetime(data2020['DateTime'])

# Choose the date you want to plot (e.g., '10-27')
selected_date = '10-27'

# Filter the DataFrames for the selected date and include only one data point per hour
selected_day_data2018 = data2018[data2018['DateTime'].dt.strftime('%m-%d') == selected_date]
selected_day_data2018 = selected_day_data2018.groupby(selected_day_data2018['DateTime'].dt.hour).first()
selected_day_data2019 = data2019[data2019['DateTime'].dt.strftime('%m-%d') == selected_date]
selected_day_data2019 = selected_day_data2019.groupby(selected_day_data2019['DateTime'].dt.hour).first()
selected_day_data2020 = data2020[data2020['DateTime'].dt.strftime('%m-%d') == selected_date]
selected_day_data2020 = selected_day_data2020.groupby(selected_day_data2020['DateTime'].dt.hour).first()

# Plot the 'Load' column for the selected day
plt.figure(figsize=(14, 5))  # Set the figure size

# Create x-axis labels with the format "dd-mm hh-mm"
x_labels = selected_day_data2018['DateTime'].dt.strftime('%d-%m %H:%M')

plt.plot(x_labels, selected_day_data2018['Load'], label='Load 2018')
plt.plot(x_labels, selected_day_data2019['Load'], label='Load 2019')
plt.plot(x_labels, selected_day_data2020['Load'], label='Load 2020')

plt.xlabel('Date and Time (dd-mm hh-mm)')
plt.ylabel('Load')
plt.title(f'Load for {selected_date}')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
